## Exploratory Data Analysis

The following analysis explores relationships between drug characteristics, approval trends, and side effect severity.

```python
df["approval_status"].value_counts()
```

This analysis examines the distribution of drug approval outcomes within the dataset. Understanding approval trends can provide insights into regulatory patterns and potential relationships between drug characteristics and approval status.

```python
df["approval_status"].value_counts().plot(kind="bar")
plt.title("Drug Approval Status Distribution")
plt.show()
```

The dataset shows variation in approval status across different drugs, indicating potential differences in safety or regulatory outcomes.
